### Using Scaler for KNN

because knn depends heavily on distance, we try to use min max scaler and keep everything between 0 and 1 to ensure no one feature dominates

always do scaling after train test split, because the scale will learn from both train and test while in real world dataset we don't know the test dataset yet and will make the test accuracy unrealistically high. Even when the test accuracy have a higher max/min score, like -0.1 or 1.05, that is perfectly fine

```python
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)

#use minmax scaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#use k=6 because it has the highest accuracy before
model = KNeighborsClassifier(n_neighbors=7)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

print("Accuracy", accuracy_score(y_test, y_pred))
```

the accuracy is still the same because the data is not that far to begin with, scaling won't do much to the accuracy

## Logistic Regression

the target is binary, try using logistic regression

```python
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)

model = LogisticRegression(max_iter = 2000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy", accuracy_score(y_test, y_pred))
```

the model is hard to converge because logistic regression works best when the feature are in the range of 0 and 1. One way is to standardized the feature. Logistic regression works best when all features are in the same scale (0-1 or z-score)

```python
data = load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter=2000)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

print("Accuracy", accuracy_score(y_test, y_pred))
```

We can see that logistic regression actually gives better accuracy than KNN. Several things why this happens:
1. Logistic Regression generalizes better
    - logistic regression learns a global relationship
    - KNN makes the decisions locally, it memorizes all training data and predicts using neighbors instead of an actual relationship between data
2. Breast cancer fits linear models well
    - the features are well behaved and linearly separable (we can roughly draw a line between benign and malignant points)
3. KNN is sensitive
    - KNN accuracy depends a lot on choosing the right k, even with scaling, while Logistic Regression doesn't have that sensitivity
    - it relies on the dataset directly and not using parameters like logistic regression. Changing the data slightly and KNN's predictions might change more than logistic regression

```python
#we use data_2 because it is a long code and not just 1 cell
#using the same variable name as before will make it messy
data_2 = sns.load_dataset('penguins')
data_2.head()
```

the data is not yet all numerical so we have to do some preprocess first.

## Preprocess

```python
data_2.island.unique()
```

```python
data_2 = data_2.dropna()
subset = data_2[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']]

subset.loc[:,'sex'] = subset['sex'].map(lambda x: 1 if x=='Male' else 0)

#unlike numerical scaling, we can use encoder before the split because categories is fixed
encoder = OneHotEncoder(sparse_output = False)
island_encode = encoder.fit_transform(data_2[['island']])
island_encode_df = pd.DataFrame(island_encode, columns = encoder.get_feature_names_out(['island']),
                               index=subset.index)

#using index=subset.index to make sure the island df has the same index as subset
#pd concat match the data with index and not order, so if indexes doesn't match then NaN appears
X = pd.concat([subset, island_encode_df], axis=1)
y = data_2['species']
```

```python
island_encode_df.head()
```

## Model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

#scaling for the numerical features
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

print("Accuracy", accuracy_score(y_test, y_pred))
```

penguins dataset is too small with only 333 rows after dropping NaN and also with the distance of the features can perfectly separate the species makes it really possible to have a 100% accuracy
